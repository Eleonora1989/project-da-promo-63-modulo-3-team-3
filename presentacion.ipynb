**Hallazgos clave (la parte central)**
Aquí va lo que ya has trabajado:
🔹 A) Horas extra → factor crítico
- 22.7% de los que NO hacen horas extra se van
- 52.7% de los que SÍ hacen horas extra se van
- Más del doble de probabilidad
Gráfico recomendado: barras + línea (conteos + tasa)

🔹 B) Antigüedad → riesgo alto en los primeros años
- Los que se van llevan menos años en la empresa
- La tasa de abandono es más alta entre 0–3 años
- Luego se estabiliza
Gráfico recomendado: tasa de attrition por years_at_company

🔹 C) Incremento salarial → NO explica la rotación
- Las medias son prácticamente iguales
- Las distribuciones se solapan
- No hay patrón en la tasa por nivel salarial
Gráfico recomendado: boxplot + KDE
