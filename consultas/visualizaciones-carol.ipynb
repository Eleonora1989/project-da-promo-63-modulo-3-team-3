```python
# Tratamiento de datos
import pandas as pd
import numpy as np

# Visualización
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración
pd.set_option('display.max_columns', None)
```

```python
# Carga de datasets

df_hr_limpio = pd.read_csv('../df_hr_limpio.csv')
```

```python
## para ver por que la gente se va se me ocurre que tendriamos que ver 


# comparar los empleados que se han ido, con los que se han quedado, para ver si hay alguna diferencia entre ellos:
# - si eran hombres o mujeres CAROL
# - si llevaban mucho tiempo con el mismo responsable o si es el mismo responsable el de todos los que se van CAROL
# - WorkLifeBalance CAROL
```

### 1) comparar los empleados que se han ido, con los que se han quedado, para ver si hay alguna diferencia entre ellos:
### - Género con mayor abandono ?

```python
plt.figure(figsize=(10,6))

sns.countplot(data=df_hr_limpio, x='attrition', hue='gender')

plt.title('Distribución de género por cantidad de empleados que se han ido o se han quedado')
plt.xlabel('Empleados que se han ido o se han quedado')
plt.ylabel('Cantidad')

plt.show()
```

### 2) comparar los empleados que se han ido, con los que se han quedado, para ver si hay alguna diferencia entre ellos:
### - si llevaban mucho tiempo con el mismo responsable o si es el mismo responsable el de todos los que se van

```python
# usaremos un barplot porque estamos viendo la relacion entre una variable categorica (attrition) y una numerica (years_with_curr_manager)
plt.figure(figsize=(10,6))

sns.barplot(data=df_hr_limpio, x='attrition', y='years_with_curr_manager')

plt.title('Empleados que se han ido y cuanto tiempo estuvieron con el mismo manager')
plt.xlabel('Empleados que se han ido o se han quedado')
plt.ylabel('Número de años con el mismo manager')



plt.show()
```

### 3) comparar los empleados que se han ido, con los que se han quedado, para ver si hay alguna diferencia entre ellos:
### - work life balance

```python
fig, axes = plt.subplots(ncols=2, nrows=1, figsize =(20,7))

sns.barplot(data=df_hr_limpio, x='attrition', y='work_life_balance', ax=axes[0])

axes[0].set_title('Nivel de Work Life Balance por empleados que se han ido o no de la empresa')
axes[0].set_xlabel('Attrition')
axes[0].set_ylabel('Work Life Balance')


# en este caso estamos viendo promedios en el barplot, ahora queremos ver distribucion en un violinplot

sns.violinplot(data=df_hr_limpio, x='attrition', y='work_life_balance', ax=axes[1])

axes[1].set_title('Nivel de Work Life Balance por empleados que se han ido o no de la empresa')
axes[1].set_xlabel('Attrition')
axes[1].set_ylabel('Work Life Balance')

plt.show()
```

### 4) comparar los empleados que se han ido, con los que se han quedado, para ver si hay alguna diferencia entre ellos:
### - Job Role con mayor abandono ?

```python
plt.figure(figsize=(10,6))

sns.countplot(data=df_hr_limpio, x='attrition', hue='job_role')

plt.title('Distribución de Job Roles por cantidad de empleados que se han ido o se han quedado')
plt.xlabel('Empleados que se han ido o se han quedado')
plt.ylabel('Cantidad')

plt.show()
```

```python
## esto podría indicar que puestos tienen burnout o una mala gestión, a continuación revisamos por departamento
```

```python
plt.figure(figsize=(10,6))

sns.countplot(data=df_hr_limpio, x='attrition', hue='department')

plt.title('Departamentos en los que empleados que se han ido o se han quedado')
plt.xlabel('Empleados que se han ido o se han quedado')
plt.ylabel('Cantidad')

plt.show()
```

### 5) comparar los empleados que se han ido, con los que se han quedado, para ver si hay alguna diferencia entre ellos:
### - Cuanto era el monthly income de los que abandonaron y de los que no ?

```python
sns.boxplot(data=df_hr_limpio, x='attrition', y='monthly_income')

plt.title('Salario Mensual de los que empleados que se han ido o se han quedado')
plt.xlabel('Empleados que se han ido o se han quedado')
plt.ylabel('Salario')


plt.show()
```

```python
## podemos observar que el promedio de salario de los que se han ido era bastante bajo, y comparandolo con los que siguen en la empresa aun mas
```

### 6) comparar los empleados que se han ido, con los que se han quedado, para ver si hay alguna diferencia entre ellos:
### - Cual era el estatus marital de los que abandonaron y de los que no ?

```python
plt.figure(figsize=(10,6))

sns.countplot(data=df_hr_limpio, x='attrition', hue='marital_status')

plt.title('Estatus marital de los que empleados que se han ido o se han quedado')
plt.xlabel('Empleados que se han ido o se han quedado')
plt.ylabel('Cantidad')

plt.show()
```

### 7) comparar los empleados que se han ido, con los que se han quedado, para ver si hay alguna diferencia entre ellos:
### - Que nivel de satisfacción tenían los que abandonaron y de los que no ?

```python
fig, axes = plt.subplots(ncols=2, nrows=1, figsize =(20,7))

sns.barplot(data=df_hr_limpio, x='attrition', y='job_satisfaction', ax=axes[0])

axes[0].set_title('Nivel de Satisfacción por empleados que se han ido o no de la empresa')
axes[0].set_xlabel('Attrition')
axes[0].set_ylabel('Nivel de Satisfacción')


# en este caso estamos viendo promedios en el barplot, ahora queremos ver distribucion en un violinplot

sns.violinplot(data=df_hr_limpio, x='attrition', y='job_satisfaction', ax=axes[1])

axes[1].set_title('Nivel de Satisfacción por empleados que se han ido o no de la empresa')
axes[1].set_xlabel('Attrition')
axes[1].set_ylabel('Nivel de Satisfacción')

plt.show()
```

### 8) comparar los empleados que se han ido, con los que se han quedado, para ver si hay alguna diferencia entre ellos:
### - Que edad tienen los que abandonaron y de los que no ?

```python
fig, axes = plt.subplots(ncols=2, nrows=1, figsize =(20,7))

sns.histplot(data=df_hr_limpio, x='age', hue='attrition', bins=30, kde=True, ax=axes[0])

axes[0].set_title('Edad de los empleados que se han ido o no de la empresa')
axes[0].set_xlabel('Edad')
axes[0].set_ylabel('Cantidad')


# en este caso estamos viendo promedios en el barplot, ahora queremos ver distribucion en un violinplot

sns.boxplot(data=df_hr_limpio, x='attrition', y='age', ax=axes[1])

axes[1].set_title('Edad de los empleados que se han ido o no de la empresa')
axes[1].set_xlabel('Empleados que se han ido o se han quedado')
axes[1].set_ylabel('Edad')

plt.show()
```
