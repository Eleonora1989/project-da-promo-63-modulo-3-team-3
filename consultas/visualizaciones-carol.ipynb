```python
# Tratamiento de datos
import pandas as pd
import numpy as np

# Visualización
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración
pd.set_option('display.max_columns', None)
```

```python
# Carga de datasets

df_hr_limpio = pd.read_csv('../df_hr_limpio.csv')
```

```python
## para ver por que la gente se va se me ocurre que tendriamos que ver 


# comparar los empleados que se han ido, con los que se han quedado, para ver si hay alguna diferencia entre ellos:
# - si eran hombres o mujeres CAROL
# - si llevaban mucho tiempo con el mismo responsable o si es el mismo responsable el de todos los que se van CAROL
# - WorkLifeBalance CAROL
```

### 1) comparar los empleados que se han ido, con los que se han quedado, para ver si hay alguna diferencia entre ellos:
### - Género con mayor abandono ?

```python
plt.figure(figsize=(10,6))

sns.countplot(data=df_hr_limpio, x='attrition', hue='gender')

plt.title('Distribución de género por cantidad de empleados que se han ido o se han quedado')
plt.xlabel('Empleados que se han ido o se han quedado')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)

plt.show()
```

### 2) comparar los empleados que se han ido, con los que se han quedado, para ver si hay alguna diferencia entre ellos:
### - si llevaban mucho tiempo con el mismo responsable o si es el mismo responsable el de todos los que se van

```python
# usaremos un barplot porque estamos viendo la relacion entre una variable categorica (attrition) y una numerica (years_with_curr_manager)
plt.figure(figsize=(10,6))

sns.barplot(data=df_hr_limpio, x='attrition', y='years_with_curr_manager')

plt.title('Empleados que se han ido y cuanto tiempo estuvieron con el mismo manager')
plt.xlabel('Empleados que se han ido o se han quedado')
plt.ylabel('Número de años con el mismo manager')


plt.xticks(rotation=45)

plt.show()
```

### 3) comparar los empleados que se han ido, con los que se han quedado, para ver si hay alguna diferencia entre ellos:
### - work life balance

```python
fig, axes = plt.subplots(ncols=2, nrows=1, figsize =(20,7))

sns.barplot(data=df_hr_limpio, x='attrition', y='work_life_balance', ax=axes[0])

axes[0].set_title('Nivel de Work Life Balance por empleados que se han ido o no de la empresa')
axes[0].set_xlabel('Attrition')
axes[0].set_ylabel('Work Life Balance')
axes[0].tick_params(axis="x", rotation=45)


# en este caso estamos viendo promedios en el barplot, ahora queremos ver distribucion en un violinplot

sns.violinplot(data=df_hr_limpio, x='attrition', y='work_life_balance', ax=axes[1])

axes[1].set_title('Nivel de Work Life Balance por empleados que se han ido o no de la empresa')
axes[1].set_xlabel('Attrition')
axes[1].set_ylabel('Work Life Balance')
axes[1].tick_params(axis="x", rotation=45)

plt.show()
```

### 4) comparar los empleados que se han ido, con los que se han quedado, para ver si hay alguna diferencia entre ellos:
### - Job Role con mayor abandono ?

```python
plt.figure(figsize=(10,6))

sns.countplot(data=df_hr_limpio, x='attrition', hue='job_role')

plt.title('Distribución de Job Roles por cantidad de empleados que se han ido o se han quedado')
plt.xlabel('Empleados que se han ido o se han quedado')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)

plt.show()
```
